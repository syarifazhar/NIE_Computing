## Question 1:
The following program searches for a list of colours to check if it contains a certain colour.

If the color is found in the list, a message is displayed on the screen that states the number of items that are in the list and position of the colour in the list. Otherwise, a message is displayed on the screen that states the colour is not in the list.

There are several syntax and logic errors in the program. Identify and correct the errors in the program. Save this file after you have completed.

```python
colour_list = ["red", "green", "blue", "orange", "purple", "yellow"]
colour_to_find = input("Which colour would you like to search for? ")

items = len(colour_list)
item_number = 0
colour_found = False

while item_number < items and not colour_found:
    if colour_list[item_number] == colour_to_find:
      colour_found = True
    else:
      item_number +=1

if colour_found:
  print("There are " + str(items) + " colours in the list, " + colour_to_find + " is item number " + str(item_number+1) + " in the list.")
else:
  print("The colour is not in the list.")




```
